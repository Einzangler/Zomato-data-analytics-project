Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
```

```python
df = pd.read_csv('zomato.csv')
df.head()
```

```python
df.shape #Analysing the size of dataset
```

```python
df.columns #analysing all the columns of dataset
```

Dropping the columns with irrelevant or unique data 

```python
df = df.drop(['url', 'address', 'phone', 'menu_item', 'dish_liked', 'reviews_list'], axis = 1)
df.head()
```

```python
df.info() #analysing the type of data stored in columns 
```

```python
df.drop_duplicates(inplace=True)
df.shape # Dropping duplicate data 
```

```python
df['rate'].unique() #cleaning the rate column 
```

```python
def modifyrate(value):   #removing "NEW" , "-" and "/5" from rate column 
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
    
df['rate'] = df['rate'].apply(modifyrate)
df['rate'].head()
```

```python
df.rate.isnull().sum()
```

```python
df['rate'].fillna(df['rate'].mean(), inplace = True) #filling null values with mean in rate column 
df['rate'].isnull().sum()
df.info()
```

```python
df.dropna(inplace=True) #dropping null values
df.head()
```

```python
df.rename(columns = {'approx_cost(for two people)':'2platescost', 'listed_in(type)':'Type'}, inplace = True) #renaming columns with long names
df.head()
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

removed listed in city column from datset-

```python
df.head()
```

```python
df['2platescost'].unique()
```

```python
def removecomma(value):
    value = str(value)
    if ',' in value :
        value = value.replace(',','')
        return float(value)
    else :
        return float(value)
    
df['2platescost']=df['2platescost'].apply(removecomma)
df['2platescost'].unique()
```

```python
df.head()
```

```python
df['rest_type'].value_counts()
```

```python
rest_types = df['rest_type'].value_counts(ascending  = False)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000] #listing restaurents with less than 1000 rest type 
rest_types_lessthan1000
```

```python
def modify_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'other restaurent'
    else:
        return value
        
df['rest_type'] = df['rest_type'].apply(modify_rest_type)
df['rest_type'].value_counts()
        
```

```python
df.head()
```

```python
df['location'].value_counts()
```

```python
location = df['location'].value_counts(ascending  = False) #clustering data in location 

location_lessthan300 = location[location<300]



def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value
        
df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
df.head()
```

```python
df['cuisines'].value_counts()
```

```python
cuisines = df['cuisines'].value_counts(ascending = False) #clustering data in cuisines 
cuisines_lessthan500 = cuisines[cuisines<500]

def modifycuisines(value):
    if(value in cuisines_lessthan500):
        return 'others'
    else :
        return value

df['cuisines'] = df['cuisines'].apply(modifycuisines)  
df['cuisines'].value_counts()
        
        
```
